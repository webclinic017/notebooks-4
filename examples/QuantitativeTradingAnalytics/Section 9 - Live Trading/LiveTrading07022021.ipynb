```python
# Raw Package
import numpy as np
import pandas as pd
```

```python
#Data Source
import yfinance as yf
```

```python
#Data viz
import plotly.graph_objs as go
```

```python
data = yf.download(tickers='BTC-USD',period = '3mo', interval = '1d')
data
```

```python
data['MA5'] = data['Close'].rolling(5).mean()
data['MA20'] = data['Close'].rolling(20).mean()
```

```python
#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

fig.add_trace(go.Scatter(x=data.index, y= data['MA20'],line=dict(color='red', width=1.5), name = 'Moving Average'))
fig.add_trace(go.Scatter(x=data.index, y= data['MA5'],line=dict(color='blue', width=1.5), name = 'Moving Average'))

fig.update_layout(
    yaxis_title='BTC/USD (kUS Dollars)')

#Show
fig.show()
```
