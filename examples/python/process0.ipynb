```python
import time
import multiprocessing

start = time.perf_counter()

def do_something(duration):
    print(f'sleeping {duration} second(s)...')
    time.sleep(duration)
    print('waking up...')

processes = []

if __name__ == '__main__':

    for _ in range(100):
        p = multiprocessing.Process(target=do_something, args=[1])
        p.start()
        processes.append(p)

    for p in processes:
        p.join()

    finish = time.perf_counter()

    print(f'Finished in {round(finish-start, 2)} second(s)')

# 100 tasks
# Finished in 3.48 second(s)
```
