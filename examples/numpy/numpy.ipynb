```python
from skimage import io

# an image is just a multidimensional array of data values
# get the image data as a numpy array
photo = io.imread('IMG_7050.jpg')
type(photo)
```

```python
# get the shapr of the image
photo.shape
```

```python
import matplotlib.pyplot as plt

# display an image
plt.imshow(photo) 
```

```python
# show image flipped vertically
p = photo[::-1]
plt.imshow(p) 
```

```python
# show image flipped horizontally
p = photo[:, ::-1]
plt.imshow(p) 
```

```python
# show a section of the image
p = photo[500:700, 850:1150]
plt.imshow(p) 
```

```python
# apply math functions to numpy array

p = np.sin(photo)
p
```

```python
print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.min(photo))
print(np.max(photo))
print(np.argmin(photo))
print(np.argmax(photo))
```

```python
z = np.array([1,2,3,4,5])

z<3 # returns arr of results of comparison
```

```python
# replace values in image
# where value is greater than 100 set to 200
# where value is less than 100 set to 0
p = np.where(photo > 100,255,0) 
plt.imshow(p)
```
